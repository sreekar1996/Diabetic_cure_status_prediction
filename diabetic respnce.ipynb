```python
features = dff[["age","gender","dbpn","sbpn","agegroup","smokingstatus","Diseasecode","hospitalsize","durationofline1","line1drugs"]]


label = dff["curestatus"]
```

```python
### Splitting data for training and test 80/20 for our Base Model - Logistic Regression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.2, random_state=0)
```
